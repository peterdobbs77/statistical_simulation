# Analysis of End of Season Rankings

```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup as bs
import matplotlib.pyplot as plt
import random
```

```python
df = pd.read_csv('./data/ultimate/archives/open_college_archive_scraped.csv')
df.head()
```

```python
top20 = df[df['Rank']<20]
print(top20)
```

```python
pvt = pd.pivot_table(df[df['Rank']<100],index='Team',columns='season',values='PR',fill_value=0)
print(pvt)
```
