# Analysis of End of Season Rankings

```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup as bs
import matplotlib.pyplot as plt
import random
```

```python
df = pd.read_csv('./data/ultimate/archives/open_college_archive_scraped.csv')
df.head()
```

```python
top20 = df[df['Rank']<=20]
pd.pivot_table(top20,index='Team',columns='season',values='PR',fill_value=0)
```

```python
schools = ['British Columbia','Brown','California-Santa Barbara','Cal Poly-SLO','Carleton College','Colorado','Florida','Georgia','Harvard','Iowa','Iowa State','Marquette','Massachusetts','Michigan','Minnesota','North Carolina','North Carolina-Wilmington','North Carolina State','Oregon','Pittsburgh','Purdue','Stanford','Texas','Texas A&M','Washington','Wisconsin','Wisconsin-Whitewater']
pd.pivot_table(df[df['Team'].isin(schools)],index='Team',columns='season',values='PR',fill_value=0)
```

```python
nc_region = ['Carleton College','Iowa','Iowa State','Marquette','Minnesota','Minnesota-Duluth','Wisconsin','Wisconsin-Milwaukee','Wisconsin-Whitewater']
pd.pivot_table(df[df['Team'].isin(nc_region)],index='Team',columns='season',values='PR',fill_value=0)
```

Looks like `Carleton College` doesn't have ranking info for 2019. This is probably because they changed their name to `Carleton College-CUT`.

```python
df[df['Team']=='Carleton College-CUT']
```

So let's adjust for that

```python
df['Team'] = df['Team'].str.replace('Carleton College-CUT','Carleton College')
nc_region = ['Carleton College','Iowa','Iowa State','Marquette','Minnesota','Minnesota-Duluth','Wisconsin','Wisconsin-Milwaukee','Wisconsin-Whitewater']
pd.pivot_table(df[df['Team'].isin(nc_region)],index='Team',columns='season',values='PR',fill_value=0)
```

Okay, that's way better. Now we're still missing data on other teams for seasons when they should have a Power Rating value. For example, check out `Wisconsin-Whitewater` for `2013_college` or `Marquette` for `2011_college`.
